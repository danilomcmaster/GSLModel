```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.cm as cm
```

```python
result_dict = {'No Intervention': {'Healthy': 0.01, 'Transitory': 0.01, 'Adverse Effects': 0.39, 'Severe Adverse Effects': 0.59}, 'Additional 250 KAF/year': {'Healthy': 0.01, 'Transitory': 0.07, 'Adverse Effects': 0.58, 'Severe Adverse Effects': 0.34}, 'Additional 800 KAF/year': {'Healthy': 0.32, 'Transitory': 0.57, 'Adverse Effects': 0.11, 'Severe Adverse Effects': 0.0}}
df = pd.DataFrame(result_dict).T
df
```

```python
df.T.plot.pie(subplots=True, figsize=(25, 10), colors=["green", "gold", "orange", "red"], labeldistance=0.5)
plt.tight_layout()
```

```python
def plot_donut_grid(df, title="", cmap="Set2"):
    n = len(df)
    ncols = 3  # adjust as needed
    nrows = (n + ncols - 1) // ncols
    
    fig, axes = plt.subplots(nrows, ncols, figsize=(5 * ncols, 5 * nrows))
    axes = axes.flatten()
    
    colors = ["Green", "Gold", "Orange", "Red"]
    
    for i, (row_name, row_data) in enumerate(df.iterrows()):
        ax = axes[i]
        wedges, texts, autotexts = ax.pie(
            row_data,
            labels=None,
            colors=colors,
            autopct=lambda p: f"{p:.1f}%" if p > 3 else "",  # hide tiny labels
            startangle=90,
            wedgeprops=dict(width=0.5, edgecolor="white", linewidth=2),  # donut hole
            pctdistance=0.75
        )
        # Style percentage text
        for autotext in autotexts:
            autotext.set_fontsize(8)
            autotext.set_color("white")
            autotext.set_fontweight("bold")
        
        ax.set_title(row_name, fontsize=13, fontweight="bold", pad=10)
    
    # Hide unused subplots
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)
    
    # Shared legend using column names
    fig.legend(
        df.columns,
        loc="lower center",
        ncol=len(df.columns),
        bbox_to_anchor=(0.5, 0),
        frameon=False,
        fontsize=10
    )
    
    fig.suptitle(title, fontsize=16, fontweight="bold", y=1.02)
    plt.tight_layout()  # leave room for legend
    plt.savefig("./images/mc_simulation_results_pie.png", bbox_inches="tight")

    plt.show()

# Usage
plot_donut_grid(df, title="10 Year Monte Carlo Simulation", cmap="Set2")
```
