```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
result_dict = {'No Intervention': {'Healthy': 0.01, 'Transitory': 0.01, 'Adverse Effects': 0.39, 'Severe Adverse Effects': 0.59}, 'Additional 250 KAF/year': {'Healthy': 0.01, 'Transitory': 0.07, 'Adverse Effects': 0.58, 'Severe Adverse Effects': 0.34}, 'Additional 800 KAF/year': {'Healthy': 0.32, 'Transitory': 0.57, 'Adverse Effects': 0.11, 'Severe Adverse Effects': 0.0}}
df = pd.DataFrame(result_dict).T
df
```

```python
df.T.plot.pie(subplots=True, figsize=(25, 10), colors=["green", "gold", "orange", "red"], labeldistance=0.5)
plt.tight_layout()
```
